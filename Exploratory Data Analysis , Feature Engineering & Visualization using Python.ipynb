#### 8. Convert the column "Reviews'' to numeric data type and check the presence of outliers in the column and handle the outliers using a transformation approach.(Hint: Use log transformation)


```python
df.dtypes   # Here as we can see 'Reviews' column has object datatype
```

```python
df['Reviews'] = df['Reviews'].astype('int')   # It is used to change the datatype of any column
```

```python
df.dtypes
```

```python
plt.figure(figsize=(15,6))   # Boxplot is used to check for the outliers in 'Reviews' column
sns.boxplot(data=df,x='Reviews')
plt.show()
```

```python
def outliers(col):   # Here we are using UDF to check for lower whisker , upper whisker and IQR
    
    q1, q3 = np.percentile(col, [25, 75])
    iqr = q3 - q1
    
    lw = q1 - 1.5 * iqr
    uw = q3 + 1.5 * iqr
    
    return lw, uw
```

```python
outliers(df['Reviews'])
```

```python
for i in ['Reviews']:
    lw, uw = outliers(df[i])
    num = ((df[i] < lw) | (df[i] > uw)).sum()
    print(f'Count of Outlier in {i} is {num}')
```

```python
sns.kdeplot(data=df,x='Reviews')   # Kdeplot gives a idea of the distribution of 'Reviews column'
plt.show()
```

```python
df['Reviews'] = np.log1p(df['Reviews'])   # Using log transformation for right skewed distribution of 'Reviews' column
```

```python
sns.boxplot(data=df,x='Reviews')   # Now outliers are handled efficiently by using log transformation
```

```python
sns.kdeplot(data=df,x='Reviews')
plt.show()
```

#### 9. The column 'Size' contains alphanumeric values, treat the non numeric data and convert the column into suitable data type. (hint: Replace M with 1 million and K with 1 thousand, and drop the entries where size='Varies with device') 

```python
df['Size'].unique()   # It gives unique values present in column 'Size'
```

```python
df = df.drop(df[df['Size']=='Varies with device'].index)
```

```python
def replace_size(value):   
    if 'M' in value:
        return float(value.replace('M', '')) * 1e6
    elif 'k' in value:
        return float(value.replace('k', '')) * 1e3
    else:
        return float(value)
```

```python
df['Size'] = df['Size'].apply(replace_size)
```

```python
df['Size'] = df['Size'].astype('float')
```
